```python
import time
```

```python
import os
os.environ["KERAS_BACKEND"] = "plaidml.keras.backend"
# os.environ["KERAS_BACKEND"] = "tensorflow"
```

```python
from keras.datasets import mnist
from keras.utils import np_utils
```

```python
# 載入minst的資料
(x_train, y_train), (x_test, y_test) = mnist.load_data()
# 將圖片轉換為一個60000*784的向量，並且標準化
x_train = x_train.reshape(x_train.shape[0], 28*28)
x_test = x_test.reshape(x_test.shape[0], 28*28)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
# try without normalization
x_train = x_train/255
x_test = x_test/255
#x_test = np.random.normal(x_test)
# 將y轉換成one-hot encoding 
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import  SGD,Adam


#建立模型
model = Sequential()
#將模型疊起 (try different units values; try adding more layers)
#relu vs. sigmoid
model.add(Dense(input_dim=28*28,units=689,activation='sigmoid'))
#model.add(Dropout(0.7))
model.add(Dense(units=689,activation='sigmoid'))
#model.add(Dropout(0.7))
model.add(Dense(units=689,activation='sigmoid'))
#model.add(Dropout(0.7))
model.add(Dense(units=10,activation='softmax'))
model.summary()
```

```python
# 記錄起始時間
start_time = time.time()
```

```python
#categorical_crossentropy vs. mse for loss
#adam vs. SGD(lr=0.1) for optimizer
model.compile(loss='categorical_crossentropy',optimizer=SGD(lr=0.1),metrics=['accuracy'])
#try different batch_size 100 vs. 10000
model.fit(x_train,y_train,batch_size=100,epochs=20)
#顯示訓練結果
score = model.evaluate(x_train,y_train)
print('\nTrain Acc:', score[1])
result = model.evaluate(x_test,y_test)
print('\nTest Acc:', result[1])
```

```python
# 記錄結束時間及執行時間
end_time = time.time()
print("--- %s seconds ---" % (end_time - start_time))
```
